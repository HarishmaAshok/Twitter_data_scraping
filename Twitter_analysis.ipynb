```python
import tweepy
import pandas as pd
from datetime import date
```

```python
consumer_key = "add your consumer_key"
consumer_secret = "add your consumer_secret"
access_token = "add your access_token"
access_token_secret = "add your access_token_secret"
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
userID = "BarackObama"

tweets = api.user_timeline(screen_name=userID,
                           count=200,
                           include_rts = True,
                           #Do not include retweets
                           # Necessary to keep full_text 
                           # otherwise only the first 140 words are extracted
                           tweet_mode = 'extended'
                           )
```

```python
tweetss = []
likes = []
timee = []
retweets = []

for info in tweets[:200]:
     
     print("ID: {}".format(info.id))
     retweets.append(info.retweet_count)
     #time_stamp = info.created_at
     timee.append(info.created_at)
     likes.append(info.favorite_count)
     tweetss.append(info.full_text)
     print("\n")
#print(len(tweetss))
#print(len(likes))
#print(len(timee))
#print(timee)
```
